### NORMALISASI

**Hasil normalisasi disimpan dalam variabel "x" untuk kemudian digunakan dalam algoritma pembelajaran mesin atau analisis data lainnya**

```python
y = data.diagnosis.values
x_data = data.drop(["diagnosis"],axis=1)
x = (x_data - np.min(x_data))/(np.max(x_data)-np.min(x_data))
x
```

# **TRAINING**

**Model Random Forest dilatih dengan data sebanyak 75%, dan sebanyak 25% dari data akan digunakan sebagai pengujian**

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25,random_state=42)
```

# **EVALUATING**

### RANDOM FOREST CLASSIFICATION

```python
# Training model
rf = RandomForestClassifier(n_estimators=10, random_state=1)
rf.fit(x_train, y_train)

# Making predictions
predictions = rf.predict(x_test)

# Evaluation metrics
print("Accuracy on Test Data Random Forest Classifier algo: ", rf.score(x_test, y_test))
rf_acc_score = rf.score(x_test, y_test)
print(f"Precision Score: {precision_score(y_test, predictions)}")
print(f"Recall Score: {recall_score(y_test, predictions)}")
print(f"F1 Score: {f1_score(y_test, predictions)}")

# Confusion matrix
cm = confusion_matrix(y_test, predictions)

# Plotting confusion matrix using Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False, square=True)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix for Test Data")
plt.show()

# Classification report
print(classification_report(y_test, predictions))

# Accuracy on Whole Data
predictions_whole = rf.predict(x.values)
accuracy_whole = accuracy_score(y, predictions_whole)
print(f"Accuracy on Whole Data: {accuracy_whole}")
print(f"Precision Score: {precision_score(y, predictions_whole)}")
print(f"Recall Score: {recall_score(y, predictions_whole)}")
print(f"F1 Score: {f1_score(y, predictions_whole)}")

# Confusion matrix for whole data
cm_whole = confusion_matrix(y, predictions_whole)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_whole, annot=True, fmt="d", cmap="Blues", cbar=False, square=True)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix for Whole Data")
plt.show()

# Classification report for whole data
print(classification_report(y, predictions_whole))
```

```python
#Random Forest Accuracy
y_pred = rf.predict(x_test)
y_true = y_test

cm = confusion_matrix(y_true, y_pred)

#visualize
f, ax = plt.subplots(figsize=(5,5))
sns.heatmap(cm,annot = True, linewidths=0.5,linecolor="red",fmt = ".0f",ax=ax)
plt.xlabel("y_pred")
plt.ylabel("y_true")
plt.show()
```

### KNN

```python
# Training model
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(x_train, y_train)

# Accuray On Test Data
print("Accuracy on Test Data K Neighbors Classifier algo: {}".format(knn.score(x_test, y_test)))
knn_acc_score = knn.score(x_test, y_test)

# Making predictions
predictions = knn.predict(x_test)

# Evaluation metrics
print(f"Precision Score: {precision_score(y_test, predictions)}")
print(f"Recall Score: {recall_score(y_test, predictions)}")
print(f"F1 Score: {f1_score(y_test, predictions)}")

# Confusion matrix
cm = confusion_matrix(y_test, predictions)

# Plotting confusion matrix using Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False, square=True)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix for Test Data")
plt.show()

# Classification report
print(classification_report(y_test, predictions))

# Accuray On Whole Data
predictions_whole = knn.predict(x.values)
accuracy_whole = accuracy_score(y, predictions_whole)
print(f"Accuracy on Whole Data: {accuracy_whole}")
print(f"Precision Score: {precision_score(y, predictions_whole)}")
print(f"Recall Score: {recall_score(y, predictions_whole)}")
print(f"F1 Score: {f1_score(y, predictions_whole)}")

# Confusion matrix for whole data
cm_whole = confusion_matrix(y, predictions_whole)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_whole, annot=True, fmt="d", cmap="Blues", cbar=False, square=True)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix for Whole Data")
plt.show()

# Classification report for whole data
print(classification_report(y, predictions_whole))
```

```python
#KNN Accuracy Visual
score_list = []
for each in range(1,15):
    knn2 = KNeighborsClassifier(n_neighbors = each)
    knn2.fit(x_train,y_train)
    score_list.append(knn2.score(x_test,y_test))

#visualize
plt.plot(range(1,15),score_list)
plt.xlabel("k values")
plt.ylabel("accuracy")
plt.savefig('plot')
plt.show()
```
